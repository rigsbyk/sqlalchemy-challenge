# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func,extract
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# # reflect the tables
Base.prepare(engine,reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

## Precipitation Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Grabbing the latest date entry
last_12 = session.query(Measurement.date,Measurement.prcp).order_by(Measurement.date.desc()).first()

# Get the first item(date) out of the tuple so it is just the date
last_date = last_12[0]

# # Calculate the date 1 year ago from the last data point in the database
last_year = dt.datetime.strptime(last_date,"%Y-%m-%d") - (dt.timedelta(days=365))
# Reformatting date to ("%Y-%m-%d")
year_ago = last_year.strftime("%Y-%m-%d")

# # Perform a query to retrieve the data and precipitation scores
data_query = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= year_ago).all()

# # Save the query results as a Pandas DataFrame 
precip_df = pd.DataFrame(data_query, columns = ["Date","Precipitation"])

# #set the index to the date column
precip_df.set_index('Date',inplace=True)

# # Sort the dataframe by date
precip_df.sort_values("Date", axis=0, ascending=True,inplace=True)

# Use Pandas Plotting with Matplotlib to plot the data
precip_df.plot(figsize = (12,9))
plt.title("Precipitation (Last 12 Months)")
plt.xlabel("Date", color ="darkblue")
plt.ylabel("Precipitation", color="darkblue")
plt.legend(loc= 'best')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
precip_df.describe()
```

## Station Analysis

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(func.count(Station.name)).all()
print(f'There are {station_count[0][0]} stations')
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
most_active = session.query(Measurement.station, func.count(Measurement.station)).group_by(Measurement.station).\
order_by(func.count(Measurement.station).desc()).all()
most_active
```

```python
# Most Active Station
print(f'The most active station is: {most_active[0][0]} with {most_active[0][1]} rows')
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?

sel = [
        func.min(Measurement.tobs),
        func.max(Measurement.tobs),
        func.avg(Measurement.tobs)]

temp_calc = session.query(*sel).filter(Measurement.station == 'USC00519281').all()

# Print out the lowest, highest, and average temperatures for station USC00519281
# NOTE: I had the print all on one line,but decided it was cleaner and easier to read as three separate lines
print(f'The lowest temperature for station USC00519281 is: {temp_calc[0][0]}')
print(f'The highest temperature for station USC00519281 is: {temp_calc[0][1]}')
print(f'The average temperature for station USC00519281 is: {round(temp_calc[0][2],1)}')

      
```

```python
# Choose the station with the highest number of temperature observations.
highest_obsv = session.query(Measurement.station, func.count(Measurement.tobs)).group_by(Measurement.station).\
    order_by(func.count(Measurement.tobs).desc()).first()
highest_obsv_data = highest_obsv[0]

# Query the last 12 months of temperature observation data for this station
last_yr_obsv = session.query(Measurement.tobs).\
    filter(Measurement.date >= year_ago).filter(Measurement.station == highest_obsv_data).all()

# Create dataframe
last_year_obsv_df = pd.DataFrame(last_yr_obsv, columns=["tobs"])

# Create histogram
last_year_obsv_df.plot.hist(bins=12)
plt.title("Station USC00519281(Last 12 Months)",color="darkblue")
plt.xlabel("Temperature",color="darkblue")
plt.ylabel("Frequency",color="darkblue")
plt.tight_layout()
plt.show()
```

### JUNE

```python
# Extract June dates
june_dates = session.query(Measurement.date,Measurement.station,func.avg(Measurement.tobs)).\
filter(extract('month',Measurement.date)=='06').group_by(Measurement.station).all()
june_dates
```

```python
# Create June DataFrame
june_population = pd.DataFrame(june_dates,columns=["date","station","temperature"])
june_population
```

### December

```python
# Extract December dates
december_dates = session.query(Measurement.date,Measurement.station,func.avg(Measurement.tobs)).\
filter(extract('month',Measurement.date)=='12').group_by(Measurement.station).all()
december_dates
```

```python
# Create December DataFrame
december_population = pd.DataFrame(december_dates,columns=["date","station","temperature"])
december_population 
```

### T-Test
* I would do a paired T-Test because the variables are measuring the same thing (temperature) for each station, but just separated by time

```python
# Create Paired T-Test
stats.ttest_rel(june_population["temperature"],december_population["temperature"])
```

### Results:
* The very small P Value signifies there is very little significant difference in the temperatures between June and December. 

## Temperature Analysis II

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
mytrip = calc_temps('2016-02-28','2016-03-05')
mytrip
```

```python
# Create dataframe
mytrip_df = pd.DataFrame(mytrip,columns=["tmin","tave","tmax"])
mytrip_df
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)


mytrip_df.plot.bar(y="tave",yerr=mytrip_df["tmax"]-mytrip_df["tmin"],figsize=(4,6),alpha=0.6,color="coral",legend=None)
plt.title("Trip Avg Temp",color="darkblue")
plt.ylabel("Temp (F)",color="darkblue")
plt.xticks(np.arange(1,1.0))
plt.ylim(0,100)
plt.show()
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip
start_date = '2016-08-23'
end_date = '2016-08-27'

# Use the start and end date to create a range of dates
trip_dates = ['2016-08-23',
              '2016-08-24',
              '2016-08-25',
              '2016-08-26',
              '2016-08-27' 
             ]

# Stip off the year and save a list of %m-%d strings
trip_dates_md = [i[5:] for i in trip_dates]

# # Loop through the list of %m-%d strings and calculate the normals for each date
normals = []

for x in trip_dates_md:
    a = {}
    a["date"] = f"{x}"
    norm = daily_normals(x)
    a["Tmin"] = norm[0][0]
    a["Tavg"] = norm[0][1]
    a["Tmax"] = norm[0][2]
    
    normals.append(a)
normals
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
normals_df = pd.DataFrame(normals)
normals_df.set_index("date")
```

```python
# Plot the daily normals as an area plot with `stacked=False`
ax = normals_df.plot.area(x="date",stacked=False,figsize=(10,6))
plt.show()
```
